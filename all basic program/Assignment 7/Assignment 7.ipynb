```python
datas = {"Date": [], "Views": [],"Artist":[]}
```

```python
def operation(datas):
    for i in range(0, len(list_npy)):
        s = n.load(list_npy[i])
        datas["Date"][len(datas["Date"]):] = s[0].tolist()
        datas["Views"][len(datas["Views"]):] = s[1].tolist()
        datas["Artist"][len(datas["Artist"]):] = [list_npy[i].split("\\")[-1].split("npy")[0][:-2]] * len(s[0])
```

```python
from concurrent.futures import ThreadPoolExecutor
```

```python
with ThreadPoolExecutor(max_workers=1) as executor:
    executor.submit(operation, datas)
```

```python
df = p.DataFrame.from_dict(datas, orient='index')
df = df.transpose()
df
```

```python
def get_df_by_date(date, d):
    DF = p.DataFrame.from_dict(d, orient='index')
    DF = DF.transpose()
    return DF[DF.Date == date]
    
def get_df_by_artist(ar, d):
    DF = p.DataFrame.from_dict(d, orient='index')
    DF = DF.transpose()
    return DF[DF.Artist == ar]
```
